```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
```

```python
# Caminho correto do dataset (o notebook corre a partir de notebooks/, os dados estão em ../data)
data_root = Path("../data/mvtec")
print("data_root:", data_root.resolve())

# Lista todas as categorias
categories = [d.name for d in data_root.iterdir() if d.is_dir()]
print("Categorias disponíveis:", categories)

# Para uma categoria, conta imagens
category = "bottle"
train_normal = list((data_root / category / "train" / "good").glob("*.png"))
print(f"\nImagens de treino normais ({category}): {len(train_normal)}")

test_folder = data_root / category / "test"
for defect_type in test_folder.iterdir():
    if defect_type.is_dir():
        imgs = list(defect_type.glob("*.png"))
        print(f"  Test - {defect_type.name}: {len(imgs)} imagens")
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

# Linha de cima: imagens normais
for i, ax in enumerate(axes[0]):
    img = Image.open(train_normal[i])
    ax.imshow(img)
    ax.set_title("Normal")
    ax.axis("off")

# Linha de baixo: defeituosas
defect_types = [d for d in test_folder.iterdir() if d.name != "good"]
for i, ax in enumerate(axes[1]):
    if i < len(defect_types):
        imgs = list(defect_types[i].glob("*.png"))
        img = Image.open(imgs[0])
        ax.imshow(img)
        ax.set_title(f"Defeito: {defect_types[i].name}")
        ax.axis("off")

plt.tight_layout()
plt.savefig("exploracao.png", dpi=150)
plt.show()
```
